```python
from scipy.stats import chi2_contingency
import numpy as np

# Contingency table: rows = tenure buckets, columns = [problem_orders, non_problem_orders]
# Using your actual numbers from segment_tenure.csv

data = {
    '0-30 days (new)':       {'total': 12253, 'problem': 132},
    '30-90 days':            {'total': 16451, 'problem': 85},
    '90+ days (established)':{'total': 83946, 'problem': 332},
}

contingency_table = []
labels = []
for bucket, vals in data.items():
    non_problem = vals['total'] - vals['problem']
    contingency_table.append([vals['problem'], non_problem])
    labels.append(bucket)

contingency_table = np.array(contingency_table)

print("Contingency Table:")
for label, row in zip(labels, contingency_table):
    print(f"  {label}: problem={row[0]}, non_problem={row[1]}")

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"\nChi-square statistic: {chi2:.2f}")
print(f"p-value: {p:.10f}")
print(f"Degrees of freedom: {dof}")

if p < 0.05:
    print("\n=> Statistically significant (p < 0.05). Seller tenure IS a real driver of cancellation rate, not noise.")
else:
    print("\n=> Not statistically significant. Difference could be due to chance.")
```
